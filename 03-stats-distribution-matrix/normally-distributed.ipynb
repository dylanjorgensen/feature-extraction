# Is The Data Normally Distributed?

It's also interesting to look at the scatter matrix with the features reordered with highly correlated features first (correlation values are in red text). The correlations really stand out with log-transformed data...

```python
%matplotlib inline
import matplotlib.pyplot as plt

idx_reorder = ['Detergents_Paper', 'Grocery', 'Milk', 'Fresh', 'Frozen', 'Delicatessen']
axes = pd.scatter_matrix(df[idx_reorder], alpha = 0.3, figsize = (14,8), diagonal = 'kde')
corr = df[idx_reorder].corr().as_matrix()

for i, j in zip(*plt.np.triu_indices_from(axes, k=1)):
    axes[i, j].annotate("%+.3f" %corr[i,j], (0.8, 0.2), xycoords='axes fraction', ha='center', va='center',color="red", fontsize=14)
```
